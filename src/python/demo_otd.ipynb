```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest

# ===========================
# 1. SimulaciÃ³n de Datos del AWR con MÃ©tricas Adicionales
# ===========================
np.random.seed(42)
# Crear un rango de fechas simuladas (Ãºltimos 30 dÃ­as, cada hora)
dates = pd.date_range(start="2025-01-01", periods=24 * 30, freq='H')

# Generar datos de session_count (sesiones activas)
session_counts = np.random.normal(loc=20, scale=5, size=len(dates))  # Normal
session_counts[::50] = session_counts[::50] * 3  # Inyectar anomalÃ­as

# Generar datos de wait_class (tipos de espera en Oracle)
wait_classes = np.random.choice(["CPU", "User I/O", "System I/O", "Concurrency", "Cluster"], size=len(dates))

# Crear mÃ©tricas adicionales para cada hora
cpu_usage = np.random.normal(loc=50, scale=10, size=len(dates))  # Uso de CPU (%)
wait_time = np.random.normal(loc=100, scale=20, size=len(dates))  # Tiempo de espera (ms)
user_io = np.random.normal(loc=50, scale=15, size=len(dates))  # User I/O (ms)
concurrency_waits = np.random.normal(loc=5, scale=2, size=len(dates))  # ContenciÃ³n
elapsed_time = np.random.normal(loc=200, scale=50, size=len(dates))  # Tiempo total de ejecuciÃ³n

# Inyectar anomalÃ­as en algunas mÃ©tricas
cpu_usage[::50] *= 2
wait_time[::60] *= 2
user_io[::70] *= 3

# Crear DataFrame con nuevas mÃ©tricas
df_extended = pd.DataFrame({
    "sample_time": dates,
    "session_count": session_counts,
    "cpu_usage": cpu_usage,
    "wait_time": wait_time,
    "user_io": user_io,
    "concurrency_waits": concurrency_waits,
    "elapsed_time": elapsed_time,
    "wait_class": wait_classes
})

display(df_extended)


# ===========================
# 2. Entrenamiento de Isolation Forest con MÃºltiples MÃ©tricas
# ===========================
features = ["session_count", "cpu_usage", "wait_time", "user_io", "concurrency_waits", "elapsed_time"]
model = IsolationForest(contamination=0.05, random_state=42)
df_extended["anomaly"] = model.fit_predict(df_extended[features])

# Identificar anomalÃ­as (outliers = -1)
df_extended["is_anomaly"] = df_extended["anomaly"] == -1

# ===========================
# 3. VisualizaciÃ³n de Resultados (AnomalÃ­as en MÃ©tricas Clave)
# ===========================
plt.figure(figsize=(14, 6))
sns.lineplot(data=df_extended, x="sample_time", y="session_count", label="Session Count", color="blue")
sns.scatterplot(data=df_extended[df_extended["is_anomaly"]], x="sample_time", y="session_count", color="red", label="AnomalÃ­a", s=50)

plt.title("DetecciÃ³n de AnomalÃ­as en AWR con Isolation Forest (MÃºltiples MÃ©tricas)")
plt.xlabel("Fecha")
plt.ylabel("NÃºmero de Sesiones Activas")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

# Mostrar los registros con anomalÃ­as detectadas
df_anomalies_extended = df_extended[df_extended["is_anomaly"]]
display(df_anomalies_extended)
```
